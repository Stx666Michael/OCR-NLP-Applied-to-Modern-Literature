```python
import csv
import numpy as np

def get_items_count(file_path):
    count_dict = {}
    with open(file_path, 'r', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            item = row[1]
            if item in count_dict:
                count_dict[item] += 1
            else:
                count_dict[item] = 1
    return count_dict

items_count_dict = get_items_count('../target/data.csv')
```

```python
sorted_dict = dict(sorted(items_count_dict.items(), key=lambda item: item[1], reverse=True))
print(sorted_dict)
print('Num of items:', len(sorted_dict))

# Remove '__' key
sorted_dict.pop('__', None)
print('Num of items after popping:', len(sorted_dict))
```

```python
# Plot the distribution of values, where the x-axis is the value and the y-axis is the count in exponential scale

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font_path = '/Users/tianxiangsong/Library/Fonts/cambria.ttf'
font_l = FontProperties(fname=font_path, size=36)
font_s = FontProperties(fname=font_path, size=24)

# Plot the distribution of values
plt.figure(figsize=(30, 10))
plt.hist(sorted_dict.values(), bins=60, color='skyblue', edgecolor='tab:blue')
plt.xlabel('Page Count', fontproperties=font_l)
plt.ylabel('Category Count', fontproperties=font_l)
plt.yscale('log')
plt.xticks(fontproperties=font_s)
plt.yticks(fontproperties=font_s)
plt.title('Distribution of page count values of all categories in Funü Zazhi', fontproperties=font_l)
plt.show()
```

```python
def get_issues_count(file_path):
    count_dict = {}
    with open(file_path, 'r', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            month = '19' + row[0][:2] + '-' + row[0][2:4]
            if month in count_dict:
                count_dict[month] += 1
            else:
                count_dict[month] = 1
    return count_dict

issues_count_dict = get_issues_count('../target/data.csv')
```

```python
# Plot the values in each year, where the x-axis is the year and the y-axis is the count

# Extract keys and values
keys = issues_count_dict.keys()
values = issues_count_dict.values()

# Convert keys to datetime64
keys = np.asarray(list(keys), dtype='datetime64[s]')

# Create a plot
plt.figure(figsize=(30, 10))
plt.plot(keys, values)

# Add labels and title
plt.xlabel('Time', fontproperties=font_l)
plt.ylabel('Page Count', fontproperties=font_l)
plt.xticks(fontproperties=font_s)
plt.yticks(fontproperties=font_s)
plt.title('Page count of each issue (month) in Funü Zazhi', fontproperties=font_l)
plt.show()
```
